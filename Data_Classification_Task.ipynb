```python
df1 = df.drop_duplicates()
```

```python
df1.duplicated().sum()
```

```python
X = df1.iloc[:,:-1].values
y  = df1.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X, y, test_size = 0.20, random_state = 4)
```

```python
print(xtrain)
```

```python
print(ytrain)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
```

```python
DT.fit(xtrain,ytrain)
```

```python
dt_pred = DT.predict(xtest)
print(dt_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
conf_matrx = confusion_matrix(ytest , dt_pred) 
print(conf_matrx)
```

# here we can see the confusion matrix with 3D array with rows and columns 
* we will calculate the accuracy score for each row predictions 

```python
#For class 0, there are 16 true positives (TP) and 0 false positives (FP), so the accuracy for class 0 
test_acc_0 = conf_matrx[0][0]/conf_matrx[0].sum()
print(test_acc_0)
```

```python
#For class 1, there are 7 true positives (TP) and 0 false positives (FP), so the accuracy for class 1
test_acc_1 = conf_matrx[1][1]/conf_matrx[1].sum()
print(test_acc_1)
```

```python
#For class 2, there are 6 true positives (TP) and 1 false positives (FP), so the accuracy for class 2
test_acc_2 = conf_matrx[2][2]/conf_matrx[2].sum()
print(test_acc_2)
```

```python
dt_classification_report = classification_report(ytest, dt_pred)
dt_accuracy = accuracy_score(ytest, dt_pred)

print(f'accuracy score for the DecisionTreeClassifier model is : {dt_accuracy}')
print('********************************************************************')
print(f'Classification Report for the DecisionTreeClassifier  model is :')
print(dt_classification_report)
```

```python
X1 = df1.iloc[:,:-1].values
y1 = df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X1, y1, test_size = 0.20, random_state = 4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNC = KNeighborsClassifier(n_neighbors =3)
KNC.fit(xtrain,ytrain)
```

```python
Knc_pred = KNC.predict(xtest)
```

```python
knn_conf_matrx = confusion_matrix(ytest , Knc_pred) 
print(knn_conf_matrx)
```

```python
KNC_classification_report = classification_report(ytest, Knc_pred)
KNC_accuracy = accuracy_score(ytest, Knc_pred)

print(f'accuracy score for the KNN model is : {KNC_accuracy}')
print('********************************************************************')
print(f'Classification Report for the KNN  model is :')
print(KNC_classification_report)
```
